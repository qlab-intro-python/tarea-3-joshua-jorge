4. Count the number of duplicates in the Google Play dataset. You must build a function to loop through the Google Play dataset

````python
# Solution

def count_dup():
    dict_duplicados = {}
    for element in list(dataset_list[1][0]['App'].values):

        if element not in dict_duplicados.keys():
            dict_duplicados[element] = 1
        elif element in dict_duplicados.keys():
            dict_duplicados[element] +=1

    total_duplicados = 0
     
    for keys, values in dict_duplicados.items():
        if values>1:
            total_duplicados += values-1

    return total_duplicados

print("Number of duplicates in the Google Play dataset:", count_dup())
````

6. Remove non-english apps from both Android and Apple datasets. You should write a function that takes a string, if there's more than three characters that fall outside the ASCII range (0 - 127), then the function should return False (identify the string as non-English), otherwise it should return True. Hint: use the [```ord()```  function](https://www.w3schools.com/python/ref_func_ord.asp)

````python
# For instance
correct_english_name = "Instachat 😜"  # 1 character outside the ASCII range (😜)
incorrect_english_name = "爱奇艺PPS -《欢乐颂2》电视剧热播" # 13 character outside the ASCII range (ie 播)
````

````python
print(ord("I"))
print(ord("t"))
print(ord("-"))
print(ord("😜"))
print(ord("爱"))
````

````python
# Solution
apple_dataset = dataset_list[0][0].drop_duplicates(subset='track_name')
google_dataset = dataset_list[1][0].drop_duplicates(subset='App')
````

````python
def english_apps(valor: str):
    contador = 0
    for element in valor:
        posicion_ord = ord(element)
        if posicion_ord >= 128:
            contador += 1
        if contador >= 4:
            return False
    return True
````

````python
apple_dataset.loc[(apple_dataset.track_name.apply(english_apps))]
````

````python
google_dataset.loc[(google_dataset.App.apply(english_apps))]
````
